# Replacing Blanks With 0 As Tenure is 0 and Total Charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

# Converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
sns.set_theme(style="whitegrid")

# Define a custom color palette
custom_palette = ['#3498db', '#e74c3c']  # Blue for 'No Churn', Red for 'Churn'

# Create the count plot and assign 'Churn' to both 'x' and 'hue'
ax = sns.countplot(x='Churn', hue='Churn', data=df, palette=custom_palette, legend=False)

# Add bar labels
ax.bar_label(ax.containers[0], label_type='edge', fontsize=12, color='black')

# Set title and axis labels
plt.title("Customer Churn Count", fontsize=16, fontweight='bold', color='#2c3e50')
plt.xlabel("Churn Status", fontsize=14, color='#2c3e50')
plt.ylabel("Number of Customers", fontsize=14, color='#2c3e50')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(3, 4))

# Group the data by 'Churn' and count occurrences
gb = df['Churn'].value_counts()

# Create a pie chart with percentage labels
plt.pie(gb, labels=gb.index, autopct="%1.2f%%", colors=['#3498db', '#e74c3c'], startangle=90, explode=[0.05, 0])

# Set aspect ratio to be equal to ensure the pie chart is circular
plt.axis('equal')

# Display the plot
plt.show()
```

# From the given pie chart we can conclude that 26.54% of our customers have churned out 
# Now let's explore the reason behind it

```python
# Set the figure size
plt.figure(figsize=(5, 4))

# Create a count plot with 'gender' on the x-axis and 'Churn' as the hue
sns.countplot(x="gender", data=df, hue="Churn", palette=['#3498db', '#e74c3c'])

# Add a title to the plot
plt.title("Churn By Gender", fontsize=16, fontweight='bold')

# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(5, 4))

# Create a count plot with 'SeniorCitizen' on the x-axis and 'Churn' as the hue
sns.countplot(x="SeniorCitizen", data=df, hue="Churn", palette=['#3498db', '#e74c3c'])

# Add a title to the plot
plt.title("Churn By Senior Citizen", fontsize=16, fontweight='bold')

# Display the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(4, 4))

# Create the count plot for 'SeniorCitizen' with both 'x' and 'hue' assigned
ax = sns.countplot(x="SeniorCitizen", hue="SeniorCitizen", data=df, palette=['#3498db', '#e74c3c'], legend=False)

# Add bar labels to display the count values
ax.bar_label(ax.containers[0], fontsize=12, color='black')

# Add a title to the plot
plt.title("Count of Customers by Senior Citizen", fontsize=14, fontweight='bold')

# Display the plot
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(5, 4)) # Adjust figsize for better visulization

# plt the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e']) # churn colors if desired

# Add percwentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title("churn by Senior Citizen (Starced Bar Chart)")
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))

plt.show()
```

```python
# List of columns for which we want to create count plots
columns = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
       'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols # calculte number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4)) # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 10 and 20 arrays)
axes = axes.flatten()

# Iteration over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Count plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

#
for j in range(i + 1,len(axes)):
    fig.delaxes(axes[j])
    
    
plt.tight_layout()
plt.show()
```
